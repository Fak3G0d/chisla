Задание 2: 1). Выбрать на плоскости 2 произвольные точки в качестве узлов и показать, как влияют значения производных в этих точках на поведение многочлена.

```python
def f_H(x, dif_a, dif_b):
    k = x - x_a
    return (1-3*k**2+2*k**3)*y_a+(3*k**2-2*k**3)*y_b+(k-2*k**2+k**3)*dif_a+(-k**2+k**3)*dif_b
```

```python
x_a = 50 * np.random.random() - 25
y_a = 50 * np.random.random() - 25
x_b = 50 * np.random.random() - 25
y_b = 50 * np.random.random() - 25
```

```python
for i in range(5):
  dif_a = 200 * np.random.random() - 100
  dif_b = 200 * np.random.random() - 100
  x_space = np.linspace(-25, 25, 100)
  y_space = [f_H(x, dif_a, dif_b) for x in x_space]
  plt.plot(x_space, y_space, label=i)
plt.legend()
plt.show()
```
