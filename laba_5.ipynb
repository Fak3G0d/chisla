Задание. По любому произвольно задаваемому набору точек построить среднеквадратическое приближение методом наименьших квадратов для следующих функций:

1. линейные функции y=a0+a1*x
2. квадратичные функции y=a0+a1*x+a2*x2
3. степенные функции y=ax^b

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10
x = np.random.random(n) * 20
y = np.random.random(n) * 5
```

```python
print(x)
print(y)
```

1. Линейные функции y=a0+a1*x

```python
xi2 = sum(x*x)
xi = sum(x)
xiyi = sum(x*y)
yi = sum(y)
left = np.array([[xi2, xi],
                 [xi, n]])
right = np.array([xiyi,
                  yi])
solved = np.linalg.solve(left, right)
x_cord = np.linspace(min(x), max(x), 50)
y_cord = [solved[0] * i + solved[1] for i in x_cord]
plt.plot(x, y, "o")
plt.plot(x_cord, y_cord)
plt.show()
```

2. Квадратичные функции y=a0+a1*x+a2*x2

В нашем случае m = 2

```python
n = 10
x = np.random.random(n) * 20
y = np.random.random(n) * 5
```

```python
c0 = n
c1 = sum(x)
c2 = sum(x*x)
c3 = sum(x*x*x)
c4 = sum(x*x*x*x)
b0 = sum(y)
b1 = sum(y*x)
b2 = sum(y*x*x)
left = np.array([[c0, c1, c2],
                 [c1, c2, c3],
                 [c2, c3, c4]])
right = np.array([b0,
                  b1,
                  b2])
solved = np.linalg.solve(left, right)
x_cord = np.linspace(min(x), max(x), 50)
y_cord = [solved[0] + solved[1]*i +solved[2]*i*i  for i in x_cord]
plt.plot(x, y, "o")
plt.plot(x_cord, y_cord)
plt.show()
```

3. Степенные функции y=ax^b

Заметим сходство с первой функцией



```python
n = 10
x = np.random.random(n) * 20
y = np.random.random(n) * 5
x_l = [float(np.log(i)) for i in x]
y_l = [float(np.log(i)) for i in y]
print(x)
print(y)
```

```python
def hand_sum(a, b = None):
    suma = 0
    if b:
        for i in range(len(a)):
            suma = suma + a[i]*b[i]
    else:
        for i in a:
            suma = suma + i
    return suma
```

```python
xi2 = hand_sum(x_l,x_l)
xi = hand_sum(x_l)
xiyi = hand_sum(x_l,y_l)
yi = hand_sum(y_l)
left = np.array([[xi2, xi],
                 [xi, n]])
right = np.array([xiyi,
                  yi])
solved = np.linalg.solve(left, right)
x_cord = np.linspace(min(x), max(x), 50)
y_cord = [np.exp(solved[0]) * i**solved[1] for i in x_cord]
plt.plot(x, y, "o")
plt.plot(x_cord, y_cord)

plt.show() 
```
